```python
import pandas as pd
from sklearn.preprocessing import LabelBinarizer
```

```python
hhtrainA=pd.read_csv('../data/raw/Pover-T_Tests_Predicting_Poverty_-_Country_A_Household_Train.csv')
hhtrainB=pd.read_csv('../data/raw/Pover-T_Tests_Predicting_Poverty_-_Country_B_Household_Train.csv')
hhtrainC=pd.read_csv('../data/raw/Pover-T_Tests_Predicting_Poverty_-_Country_C_Household_Train.csv')

hitrainA=pd.read_csv('../data/raw/Pover-T_Tests_Predicting_Poverty_-_Country_A_Individual_Train.csv')
hitrainB=pd.read_csv('../data/raw/Pover-T_Tests_Predicting_Poverty_-_Country_B_Individual_Train.csv')
hitrainC=pd.read_csv('../data/raw/Pover-T_Tests_Predicting_Poverty_-_Country_C_Individual_Train.csv')
```

# Consolidando en una sola tabla por país

Hay que generar modelos por país, dadas las diferencias en las preguntas de las encuestas por país

```python
train_A=pd.merge(hitrainA,hhtrainA,how='inner',on='id')
train_A.head()
```

```python
train_B=pd.merge(hitrainB,hhtrainB,how='inner',on='id')
train_B.head()
```

```python
train_C=pd.merge(hitrainC,hhtrainC,how='inner',on='id')
train_C.head()
```

# Codificando las variables categóricas

```python
def binarize_df(df_in):
    cols = []
    for i in df_in.columns:
        if df_in[i].dtype == 'object':
            cols.append(i)

    df = pd.DataFrame()
    for i in cols:
        lb = LabelBinarizer()
        column_binarized = pd.DataFrame(lb.fit_transform(df_in[i]))
        if len(column_binarized.columns) < 2:
            column_binarized.columns = [i + '_' + lb.classes_[0]]
        else:
            column_binarized.columns = lb.classes_
            column_binarized.columns = [
                i + '_' + str(col) for col in column_binarized.columns
            ]
        df = pd.concat(
            [
                df,
                column_binarized,
            ], axis=1)
        # Faltan agregar las variables no categóricas al dataset
    return df
```

```python
train_A_binarized = binarize_df(train_A)
train_A_binarized.to_csv('../data/interim/train_A_binarized.csv')
train_B_binarized = binarize_df(train_B)
train_B_binarized.to_csv('../data/interim/train_B_binarized.csv')
train_C_binarized = binarize_df(train_C)
train_C_binarized.to_csv('../data/interim/train_C_binarized.csv')
```
